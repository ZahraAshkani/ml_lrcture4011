## Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing as prep
from scipy.stats import variation
```

## Loading dataset

```python
df_iris = pd.read_csv('Data/iris.txt', sep= ',', header= None) 
df_iris.head()
```

## Renaming of headers and encoding 

```python
df_iris.columns= ['sepal_l', 'sepal_w', 'petal_l', 'petal_w', 'iris_name']
encoder = LabelEncoder()
classes = encoder.fit_transform(df_iris['iris_name'])
df_iris['class'] = classes
df_iris.head()
```

## STD

## Covariance 

```python
df_iris.cov()
```

## Correlation

```python
df_iris.corr()
```

## total variance & generalized variance

```python
cov_matrix = df_iris.cov()
tot_var = np.round(np.trace(cov_matrix),5)
print('total variance =',tot_var)
gen_var = np.round(np.linalg.det(cov_matrix),10)
print('generalized variance =',gen_var)
```
